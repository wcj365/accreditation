```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IACBE.csv")
df.head(30)
```

```python
df["Country"] = df["Location"].apply(lambda x: x if len(x) > 2 else "US")
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
df2 = df[(df["Country"] != "US")].groupby("Country")["Institution Name"].count().sort_values()
df2.plot.barh(ax=ax)
plt.savefig("IACBE.png", bbox_inches="tight")
plt.close(fig)
```

```python
acbsp = pd.read_csv("ACBSP.csv")
acbsp.head()
```

```python
fig, ax = plt.subplots(figsize=(16,10))
acbsp2 = acbsp[(acbsp["Country"] != "United States")].groupby("Country")["Institution"].count().sort_values()
acbsp2.plot.bar(ax=ax)
plt.savefig("ABCSP.png", bbox_inches="tight")
plt.close(fig)
```
